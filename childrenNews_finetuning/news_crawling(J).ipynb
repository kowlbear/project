```python
import pandas as pd

# Excel 파일 읽기
file_path = "./빅카인즈_중앙일보_1.xlsx"
raw_df = pd.read_excel(file_path, sheet_name=0)

# 필요한 열 선택 (일자, 언론사, 제목, URL)
raw_df2 = raw_df[['일자', '언론사','사건/사고 분류1','사건/사고 분류2','사건/사고 분류3', '제목', 'URL']]

# '범죄'를 포함하는 행 삭제
columns_to_check = ['사건/사고 분류1', '사건/사고 분류2', '사건/사고 분류3']
for column in columns_to_check:
    raw_df2 = raw_df2[~raw_df2[column].str.contains('범죄').fillna(False)]
```

```python
raw_df2.drop(['사건/사고 분류1','사건/사고 분류2','사건/사고 분류3'], axis=1, inplace=True)
raw_df2.head(1)
```

```python
from bs4 import BeautifulSoup
import requests

# kmib_df 데이터프레임에서 첫 번째 URL 선택
url = raw_df2['URL'].iloc[0]

# URL로부터 HTML 내용 가져오기 (EUC-KR 인코딩)
response = requests.get(url)
html_content = response.content

# HTML을 BeautifulSoup를 사용하여 파싱
news = BeautifulSoup(html_content, 'html.parser')

# 출력
#print(news)
```

```python
# BeautifulSoup를 사용하여 HTML에서 특정 요소 선택 후 텍스트 추출
selected_node = news.find("div", class_="article_body")
if selected_node is not None:
    if selected_node.find("p",class_="caption") is not None:
        selected_node.find("p",class_="caption").decompose()
    text = selected_node.get_text()
else:
    text=''
text
```

```python
# 패턴 제거
import re
def cleaned_text(text):
    after_text = re.sub("[\r\n\t]", "", text)
    return after_text
print(cleaned_text(text))
```

```python
from bs4 import BeautifulSoup
import requests
import re
import time

contents = []

for i, url in enumerate(raw_df2['URL']):
    print(f"{i + 1}: {url}")

    response = requests.get(url)
    html_content = response.text

    news = BeautifulSoup(html_content, 'html.parser')
    selected_node = news.find("div", class_="article_body")
    if selected_node is not None:
        if selected_node.find("p",class_="caption") is not None:
            selected_node.find("p",class_="caption").decompose()
        text = selected_node.get_text()
        contents.append(cleaned_text(text))
    else:
        contents.append('')

    time.sleep(0.01)
```

```python
import pandas as pd

# '본문' 열에 news_all 리스트 내용 할당
raw_df2['본문'] = contents

# CSV 파일로 저장
raw_df2.to_csv('빅카인즈 기사(중앙1).csv', index=False, encoding='utf-8-sig')
```
