```python
import pandas as pd

# pandas 표시 옵션 조정
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

# Excel 파일 불러오기
df = pd.read_csv('./빅카인즈 기사_통합본_중앙.csv')

# 각 행을 개별 텍스트 파일로 저장
for index, row in df.iterrows():
    # 파일 이름을 8자리 숫자 형식으로 지정
    filename = f'{index+1:08}.txt'

    # 각 열을 문자열로 변환하고 결합
    row_text = ' '.join(str(item) + ' ' for item in row).strip()
    # 텍스트 파일로 저장 (UTF-8 인코딩 사용)
    with open(filename, 'w', encoding='utf-8') as file:
    # '제목'과 '본문' 열의 데이터 추출 및 결합
        title = row['제목']
        content = row['본문']
        row_text = f'{title}\n{content}'
        file.write(row_text)
```
